```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Cropping image

```python
path_1 = "C:/Users/HH/OneDrive - Hanoi University of Science and Technology/Downloads/_MHU6804.JPG"

image = cv2.imread(path_1)
resized_image = cv2.resize(image, (700, 650))
cv2.imshow("resized", resized_image)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
# cropped = img[start_row:end_row, start_col:end_col]
img = resized_image[30:500, 200:600] 
cv2.imshow("cropped", img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

## Dividing an Image Into Small Patches Using Cropping

```python
height = img.shape[0]
width = img.shape[1]
height_size = int(height/4)
width_size = int(width/4)

for y in range(0, height, height_size):
    for x in range(0, width, width_size):
        x_1 = x + width_size
        y_1 = y + height_size

        # cv2.rectangle(image, start_point, end_point, color, thickness) 
        
        if x_1 >= width and y_1 >= height:
            crop = img[y: height, x: width]
            cv2.imwrite('image' + str(x) + '_' + str(y) + '.jpg', crop)
            cv2.rectangle(img, (x, y), (width, height), (0, 255, 0), 2)
        elif x_1 >= width_size:
            crop = img[y: y_1, x: width]
            cv2.imwrite('image' + str(x) + '_' + str(y) + '.jpg', crop)
            cv2.rectangle(img, (x, y), (width, y_1), (0, 255, 0), 2)
        elif y_1 >= height:
            crop = img[y: height, x: x_1]
            cv2.imwrite('image' + str(x) + '_' + str(y) + '.jpg', crop)
            cv2.rectangle(img, (x, y), (x_1, height), (0, 255, 0), 2)
        else:
            crop = img[y: y_1, x: x_1]
            cv2.imwrite('image' + str(x) + '_' + str(y) + '.jpg', crop)
            cv2.rectangle(img, (x, y), (x_1, y_1), (0, 255, 0), 2)

# plt.imshow(img[:, :, ::-1])
plt.imshow(img[:, :, ::-1])
plt.axis('off')
cv2.imwrite("image_divided.jpg", img)
```
